```python
with open('judgment.txt', 'r') as file:
    data = file.read()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 700, #글자수
    chunk_overlap  = 140,
    length_function = len,
)
```

```python
texts = text_splitter.split_text(data)
texts
```

```python
print(len(texts))
```

```python
print(len(texts[0]))
print(texts[0])
```
