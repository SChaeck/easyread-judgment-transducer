```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
```

```python
# 웹 드라이버 초기화 (크롬 드라이버 사용 예시)
driver = webdriver.Chrome()

# 해당 URL로 이동
url = "https://elaw.klri.re.kr/kor_service/newWordSearch.do"
driver.get(url)

# 페이지 로드 대기
wait = WebDriverWait(driver, 10)

# 페이지 번호를 저장할 변수
page_number = 1

# 데이터를 저장할 리스트
all_data = []
```

```python
while True:
    try:
        # 현재 페이지의 ul 요소를 찾음
        # `iframe`으로 전환하기
        iframe = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/div[3]/form/section/iframe")))
        driver.switch_to.frame(iframe)

        # `iframe` 내부의 요소 선택하기
        ul_element = wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/ul")))
        li_elements = ul_element.find_elements(By.TAG_NAME, "li")

        for li in li_elements:
            print("sadfsaf")
            a_tag = li.find_element(By.TAG_NAME, "a")
            print("a_tag:", a_tag.text)
            a_tag.click()

            # `iframe`에서 기본 문서로 전환
            driver.switch_to.default_content()

            # 필요한 데이터 수집
            try:
                td1 = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/div[3]/section/table/tbody/tr[2]/td[1]")))
                td3 = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/div[3]/section/table/tbody/tr[2]/td[3]")))
                all_data.append([td1.text, td3.text])
                print(all_data)
            except:
                print("Required elements not found on the page.")
                continue
            
            # 다시 `iframe`으로 전환
            iframe = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/div[3]/form/section/iframe")))
            driver.switch_to.frame(iframe)
            
            # `iframe` 내부의 요소 선택하기 (다시 탐색)
            ul_element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/ul")))
            li_elements = ul_element.find_elements(By.TAG_NAME, "li")
            print("Re-found li elements:", li_elements)

        # `iframe`에서 기본 문서로 전환
        driver.switch_to.default_content()


        # 다음 페이지로 이동
        next_buttons = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div[3]/form/section/div/div[2]/button")
        if next_buttons:
            for btn in next_buttons:
                if btn.text == str(page_number + 1):
                    btn.click()
                    time.sleep(2)
                    page_number += 1
                    break
            else:
                # 다음 페이지 버튼을 찾지 못한 경우
                print("No more pages to navigate.")
                break
        else:
            # 다음 페이지 버튼이 없는 경우
            print("No more pages to navigate.")
            break

    except TimeoutException:
        print("Timeout occurred waiting for the page to load")
        break
    except Exception as e:
        print(f"An error occurred: {e}")
        break
```

```python
for data in all_data:
    print(data[0].split("\n"))
```

```python
import csv
# 수집한 데이터 출력
extracted_data = [data[0].split("\n") for data in all_data]
print(extracted_data)

# CSV 파일 저장
csv_filename = "extracted_data.csv"
with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # CSV 파일에 헤더 추가 (필요시)
    writer.writerow(['name', 'meaning'])
    writer.writerows(extracted_data)

print(f"Data has been written to {csv_filename}")
```

```python
# 드라이버 종료
driver.quit()
```
